```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('/content/imigrantes_canada.csv')
```

```python
df
```

```python
df.info()
```

```python
df.set_index('País', inplace=True)
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
brasil = df.loc['Brasil', anos]
brasil
```

```python
brasil_dict = {'ano': brasil.index.to_list(), 'imigrantes': brasil.values.tolist()}
dados_brasil = pd.DataFrame(brasil_dict)
dados_brasil
```

```python
plt.figure(figsize=(8, 4))
plt.plot(dados_brasil['ano'], dados_brasil['imigrantes'])
anos = np.arange(1980, 2014).astype(str)
plt.xticks(anos)
plt.title('Imigrantes do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Imigrantes')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax.plot(dados_brasil['ano'], dados_brasil['imigrantes'], lw=3, color='g')
ax.set_title('Imigrantes do Brasil para o Canadá\n 1980 a 2013', fontsize=18, loc='left')
ax.set_xlabel('Ano', fontsize=14)
ax.set_ylabel('Imigrantes', fontsize=14)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.grid()
fig.savefig('imigrantes_brasil.png',bbox_inches='tight', dpi=300)
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
axs[0].plot(dados_brasil['ano'], dados_brasil['imigrantes'])
axs[0].set_title('Imigrantes do Brasil para o Canadá')
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Imigrantes')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].grid()
axs[1].boxplot(dados_brasil['imigrantes'])
axs[1].set_title('Boxplot dos Imigrantes do Brasil para o Canadá')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('Imigrantes')
axs[1].grid()
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10,6))
fig.subplots_adjust(hspace=0.5, wspace=0.3)
fig.suptitle('Imigrantes para o Canadá da america do sul de 1980 a 2013')
axs[0,0].plot(df.loc['Brasil', anos])
axs[0,0].set_title('Brasil')
axs[0,1].plot(df.loc['Colômbia', anos])
axs[0,1].set_title('Colômbia')
axs[1,0].plot(df.loc['Argentina', anos])
axs[1,0].set_title('Argentina')
axs[1,1].plot(df.loc['Chile', anos])
axs[1,1].set_title('Chile')

ymin = 0
ymax = 7000

for ax in axs.flat:
    ax.set(xlabel='Ano', ylabel='Imigrantes')
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.grid()
    ax.set_ylim(ymin, ymax)

plt.show()
```

```python
america_sul = df.query('Região=="América do Sul"')
america_sul
```

```python
america_sul_sorted = america_sul.sort_values(by='Total', ascending=True)
america_sul_sorted
```

```python
cores = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal' , 'tomato']

fig, ax = plt.subplots(figsize=(12,5))
ax.barh(america_sul_sorted.index, america_sul_sorted['Total'], color= cores)
ax.set_title('Imigrantes para o Canadá da América do Sul de 1980 a 2013', loc='left', fontsize=18)
ax.set_xlabel('Número de Imigrantes', fontsize=14)
ax.set_ylabel('')
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

for i,v in enumerate(america_sul_sorted['Total']):
    ax.text(v+20, i, str(v), color='black', va='center', ha='left', fontsize=10)
ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='both',which='both', length=0)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))
ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h', palette='rocket')
ax.set_title('Países com maior Imigração para o Canadá \n 1980 a 2013', loc='left',fontsize=16)
ax.set_xlabel('Número de Imigrantes', fontsize=14)
ax.set_ylabel('')

plt.show()

```

```python
def gerar_grafico_paleta(pallete):
    fig, ax = plt.subplots(figsize=(8,4))
    ax = sns.barplot(data=top_10, y=top_10.index, x='Total', orient='h', palette=pallete)
    ax.set_title('Países com maior Imigração para o Canadá \n 1980 a 2013', loc='left',fontsize=18)
    ax.set_xlabel('Número de Imigrantes', fontsize=14)
    ax.set_ylabel('')
    plt.show()
```
